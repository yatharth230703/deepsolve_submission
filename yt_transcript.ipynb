```python
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
from fpdf import FPDF

def fetch_transcript(video_id):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        return ' '.join([entry['text'] for entry in transcript])
    except Exception as e:
        print(f"Could not fetch transcript for video ID {video_id}: {e}")
        return None

def fetch_video_details(video_url):
    try:
        yt = YouTube(video_url)
        title = yt.title
        transcript = fetch_transcript(yt.video_id)
        return title, transcript
    except Exception as e:
        print(f"Could not fetch details for video URL {video_url}: {e}")
        return None, None

def save_to_pdf(video_details, pdf_filename):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    for title, transcript in video_details:
        if transcript:
            pdf.set_font("Arial", style='B', size=14)
            pdf.cell(0, 10, title, ln=True)
            pdf.set_font("Arial", size=12)
            pdf.multi_cell(0, 10, transcript)
            pdf.ln(10)  # Add a line break between videos

    pdf.output(pdf_filename)

def main(video_urls, pdf_filename):
    video_details = []
    for video_url in video_urls:
        title, transcript = fetch_video_details(video_url)
        if title and transcript:
            video_details.append((title, transcript))

    save_to_pdf(video_details, pdf_filename)
    print(f"Saved video details to {pdf_filename}")

if __name__ == "__main__":
    # List of YouTube video URLs
    video_urls = [
        'https://www.youtube.com/watch?v=TX9qSaGXFyg',
        'https://www.youtube.com/watch?v=gnPNwzus95w'
    ]
    
    pdf_filename = 'youtube_videos.pdf'
    main(video_urls, pdf_filename)
```

```python
import os
import faiss
import openai
from langchain_community.document_loaders import UnstructuredURLLoader, UnstructuredPDFLoader
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings
from langchain.chains import RetrievalQA
from langchain.llms import OpenAI
from langchain.memory import ConversationBufferMemory

# Set up the Unstructured API key
os.environ['UNSTRUCTURED_API_KEY'] = 'your_unstructured_api_key_here'
openai.api_key = os.getenv('OPENAI_API_KEY')

# Function to load documents from URLs
def load_documents_from_urls(urls):
    url_loader = UnstructuredURLLoader(urls=urls)
    url_documents = url_loader.load()
    return url_documents

# Function to load documents from PDF files
def load_documents_from_pdfs(pdf_paths):
    pdf_loader = UnstructuredPDFLoader(filepaths=pdf_paths)
    pdf_documents = pdf_loader.load()
    return pdf_documents

# Initialize the OpenAI embeddings
embeddings = OpenAIEmbeddings()

# Function to create the FAISS vector store
def create_vector_store(documents):
    texts = [doc['text'] for doc in documents]
    vectors = embeddings.embed_documents(texts)
    
    # Create a FAISS index
    index = faiss.IndexFlatL2(len(vectors[0]))
    index.add(vectors)
    
    return FAISS(index=index, embeddings=embeddings)

# Load documents from URLs and PDFs
urls = ['http://example.com']
pdf_paths = ['path/to/pdf1.pdf', 'path/to/pdf2.pdf']
url_documents = load_documents_from_urls(urls)
pdf_documents = load_documents_from_pdfs(pdf_paths)

# Combine documents
all_documents = url_documents + pdf_documents

# Create vector store
vector_store = create_vector_store(all_documents)

# Initialize the OpenAI language model
llm = OpenAI(model="text-davinci-003", openai_api_key=os.getenv('OPENAI_API_KEY'))

# Create a memory object
memory = ConversationBufferMemory()

# Create the RetrievalQA chain with memory
qa_chain = RetrievalQA(llm=llm, retriever=vector_store.as_retriever(), memory=memory)

# Function to generate responses using the RAG system
def generate_response(query):
    response = qa_chain.run(query)
    return response

# Chatbot interface
def chatbot():
    print("Welcome to the RAG Sales Chatbot! Ask your questions below (type 'exit' to stop):")
    while True:
        query = input("You: ")
        if query.lower() == 'exit':
            break
        response = generate_response(query)
        print(f"Chatbot: {response}")

if __name__ == "__main__":
    chatbot()
```
